```python
import pandas as pd

import datetime
```

```python
df = pd.read_csv('Raw_data.csv')

pd.set_option('display.max_colwidth', None)
```

```python
df
```

```python
df['column'] = list(df['IDPerson#NameFirst#NameLast#ZIPZIP#PHONEPhone#IDCallID#TimeIME'].str.split('#'))
```

```python
df['column']
```

```python
col_list = ['ID', 'First Name', 'Last Name', 'Zip', 'Phone No', 'Call ID', 'datetime']
```

```python
num_list = list(range(7))
```

```python
num_list
```

```python
for i, j in zip(col_list, num_list):
    
    df[i] = df['column'].apply(lambda x: x[j])
```

```python
df
```

```python
df.drop(['IDPerson#NameFirst#NameLast#ZIPZIP#PHONEPhone#IDCallID#TimeIME', 'column'], axis = 1, inplace = True)
```

```python
df['First Name'] = df['First Name'].apply(lambda x: x[1:])
```

```python
df['Last Name'] = df['Last Name'].apply(lambda x: x.replace(x[x.index('('):x.index(')')+1], ''))
```

```python
df['Zip'] = df['Zip'].apply(lambda x: x.replace('???', '0'))
```

```python
df['datetime'] = df['datetime'].apply(lambda x: x.replace('_', '/'))
```

```python
df['datetime'] = df['datetime'].apply(lambda x: pd.to_datetime(x))

df['datetime'] = df['datetime'].apply(lambda x: x.strftime("%#m/%#d/%Y %H:%M"))
```
